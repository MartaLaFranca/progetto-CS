```python
ranks['world_rank_times']=ranks['world_rank_times'].str.replace('\=', '', regex=True)
ranks['world_rank_shanghai']=ranks['world_rank_shanghai'].str.replace('\=', '', regex=True)
#ranks['world_rank_cwur']=ranks['world_rank_cwur'].str.replace('^\=', '', regex=True)
```

```python
range_re=re.compile("(\d+)-(\d+)")     #assegna per ricordare
ranks['ranges_times']=ranks["world_rank_times"][ranks['world_rank_times'].str.match(range_re)]
ranks
```

```python
range_re=re.compile("(\d+)-(\d+)")     #assegna per ricordare
ranks['ranges_shanghai']=ranks["world_rank_shanghai"][ranks['world_rank_shanghai'].str.match(range_re)]
ranks
```

```python
ranks['range1']=ranks['ranges_times'].str.replace('(\d+)-(\d+)', r"\1", regex=True)
ranks['range2']=ranks['ranges_times'].str.replace('(\d+)-(\d+)', r"\2", regex=True)
```

```python
ranks["world_rank_times"]=np.where(ranks["ranges_times"].isnull(),ranks["world_rank_times"],(pd.to_numeric(ranks['range2'])+pd.to_numeric(ranks['range1']))/2)
```

```python
ranks['range1']=ranks['ranges_shanghai'].str.replace('(\d+)-(\d+)', r"\1", regex=True)
ranks['range2']=ranks['ranges_shanghai'].str.replace('(\d+)-(\d+)', r"\2", regex=True)
```

```python
ranks["world_rank_shanghai"]=np.where(ranks["ranges_shanghai"].isnull(),ranks["world_rank_shanghai"],(pd.to_numeric(ranks['range2'])+pd.to_numeric(ranks['range1']))/2)
```

```python
ranks
```

```python
ranks.drop(['ranges_times', 'ranges_shanghai',"range2",'range1'],axis=1,inplace=True)
```

```python
ranks['world_rank_shanghai']=pd.to_numeric(ranks['world_rank_shanghai'])
ranks['world_rank_times']=pd.to_numeric(ranks['world_rank_times'])
```

```python
ranks["max_diff_rank"]=ranks.max(axis=1)-ranks.min(axis=1)
```
